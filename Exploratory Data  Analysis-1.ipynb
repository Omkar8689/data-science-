```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('winequality-red (1).csv')


print(df.head())

```

```python
print(df.describe())

```

```python
plt.figure(figsize=(12, 8))
df.hist(bins=20, color='skyblue', edgecolor='black', linewidth=1.2)
plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(12, 8))
for column in df.columns:
    sns.kdeplot(df[column], fill=True, label=column)
plt.legend()
plt.show()
```
